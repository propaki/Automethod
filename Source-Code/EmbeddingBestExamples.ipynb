```python
def get_embedding(text):
    """Generate an embedding for the given text using BERT."""
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Use the mean of the last hidden states as the embedding
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings
```

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
import pickle
import os
def get_embedding(text):
    """Generate an embedding for the given text using BERT."""
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Use the mean of the last hidden states as the embedding
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

def save_embeddings(csv_path, save_dir):
    # Ensure the save directory exists
    os.makedirs(save_dir, exist_ok=True)

    # Load the CSV file
    df = pd.read_csv(csv_path)

    # Add a new column with numbering
    df['Number'] = range(1, len(df) + 1)

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        # Generate embedding for the functional description
        embedding = get_embedding(row['Functional Description'])

        # Convert the embedding to a numpy array for saving
        embedding_np = embedding.numpy()

        # Define the file name based on the row number
        file_name = os.path.join(save_dir, f'embedding_{row["Number"]}.pkl')

        # Save the embedding to a pickle file
        with open(file_name, 'wb') as file:
            pickle.dump(embedding_np, file)

# Specify the path to your CSV file
csv_path = '/content/drive/MyDrive/Method/EnglishMethod/Score-Examples/Chinese_Best_Examples/Embedded_Chinese/Best_Shots.csv'
# Specify the directory where you want to save the embeddings
save_dir = '/content/drive/MyDrive/Method/EnglishMethod/Score-Examples/Chinese_Best_Examples/Embedded_Chinese'
save_embeddings(csv_path, save_dir)
```
